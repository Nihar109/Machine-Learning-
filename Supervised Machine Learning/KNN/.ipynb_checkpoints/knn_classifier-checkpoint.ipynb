```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df.columns
```

```python
df1 = df.loc[:,['Age', 'EstimatedSalary', 'Purchased']]
df1
```

```python
data = df.iloc[:,2:]
data
```

```python
X = df.iloc[:,2:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train[0:5]
```

```python
X_test = scaler.transform(X_test)
X_test[0:5]
```

```python
# 1st method
k = np.sqrt(X_train.shape[0])
k
```

```python
k = 17 # for reduce the ambiguity
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# 2nd method

accuracy = []

for i in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    accuracy.append(accuracy_score(y_test,y_pred))
```

```python
accuracy
```

```python
plt.plot(range(1,26),accuracy)
```

```python
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# Effect of train error and validation error

error_train = []
error_test = []

for i in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    x = confusion_matrix(y_train,knn.predict(X_train))
    y = confusion_matrix(y_test,knn.predict(X_test))
    
    error_train.append((x[0][1]+x[1][0])/x.sum())
    error_test.append((y[0][1]+y[1][0])/y.sum())
```

```python
plt.plot(range(1,26),error_train,label='Training Error Rate')
plt.plot(range(1,26),error_test,label='Testing Error Rate')

plt.xlabel('K value')
plt.ylabel('Error')
plt.legend()
```

# Effect of k on decision boundary

```python
def keffect(data,k):
    data = pd.read_csv(f'toy_datasets/{data}')
    X = data.iloc[:,:2].values
    y = data.iloc[:,-1].values
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X,y)
    
    a = np.arange(X[:,0].min()-1,X[:,0].max()+1,0.01)
    b = np.arange(X[:,1].min()-1,X[:,1].max()+1,0.01)
    XX,YY = np.meshgrid(a,b)
    
    input_array = np.array([XX.ravel(),YY.ravel()]).T
    labels = knn.predict(input_array)
    
    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha=0.75)
    plt.scatter(X[:,0],X[:,1],c=y)
    
```
