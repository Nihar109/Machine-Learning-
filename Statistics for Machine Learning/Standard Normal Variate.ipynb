# 1. Standardization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
print('Import libraries sucessfully !!')
```

```python
from scipy.stats import norm

# Generate 1000 normal random integer with specified mean and std.
draw = norm.ppf(np.random.random(100000), loc=1000, scale=200).astype(int)
sns.distplot(draw)
plt.show()
```

```python
z = []

for i in draw:
    z.append((i-1000)/200)
    
z
```

```python
a = np.array(z)
```

```python
print(a.mean())
```

```python
print(np.round(a.mean()))
```

```python
print(a.std())
```

```python
print(np.round(a.std()))
```

```python
sns.distplot(a)
plt.show()
```
